```python
import numpy as np
```

```python
# Create a 1D vector with NumPy

vector_1D = np.array([-1.11, 1.96, -0.70, -0.23, 4.42])

print(vector_1D)
```

```python
# Create a 2D matrix with NumPy

matrix_2D = np.array(
    [
        [0.9, 0.72, 1.0, 1.5],
        [-1.1, 0.37, 0.69, 5.33],
        [0.97, 8.62, 1.435, 9.32]
        
    ]
)

print(matrix_2D)
```

```python
# Create Zeros vector
vector_zeros = np.zeros(4)
print(vector_zeros)
print()

# Create Zeros matrix
matrix_zeros = np.zeros((4,5))
print(matrix_zeros)
```

```python
# Create Ones vector
vector_ones = np.ones(5)
print(vector_ones)
print()

# Create Ones matrix
matrix_ones = np.ones((6,4))
print(matrix_ones)
```
