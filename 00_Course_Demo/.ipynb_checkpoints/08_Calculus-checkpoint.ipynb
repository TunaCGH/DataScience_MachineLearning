```python
import sympy as syp
import numpy as np

x, y, z = syp.symbols('x y z')
```

# <span style="color:brown;">LIMITS</span>

```python
syp.limit(
    e = syp.sin(x)/x, 
    z = x, 
    z0 = 0
)
```

```python
syp.limit(
    e = 1/x,
    z = x,
    z0 = 0,
    dir = '-' # left-handed limit
)
```

```python
syp.limit(
    e = 1/x,
    z = x,
    z0 = 0,
    dir = '+' # right-handed limit
)
```

# <span style="color:brown;">DERRIVATIVES</span>

```python
syp.diff(syp.cos(x), x)
```

```python
syp.diff(syp.exp(x**2), x) # e^(x^2)
```

```python
syp.diff(x**4, x, 2) # second order derrivative
```

```python
syp.diff(2*x**2 + y + x, x) # parital derrivative respects to x
```

```python
syp.diff(2*x**2 + y + x, y) # parital derrivative respects to y
```
