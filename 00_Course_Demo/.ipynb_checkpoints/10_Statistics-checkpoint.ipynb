```python
import polars as pl
from polars import col as c
import plotnine as pln
from scipy import stats

pl_pokemon = (
    pl.read_csv(
        source = "data/pokemon.csv",
        schema_overrides = {
            "Type 1": pl.Categorical,
            "Type 2": pl.Categorical,
            "Generation": pl.Categorical,
            "Legendary": pl.Boolean
        }
    )
    .drop("#")
    .rename(lambda col: col.strip().replace(" ", "_").replace(".", ""))
    .with_columns(Generation = c("Generation").cast(pl.Utf8).cast(pl.Enum(["1", "2", "3", "4", "5", "6"])))
)
```

```python
pl_pokemon.head(5)
```

# <span style="color:brown;">Correlation between 2 quantitative variables</span>

```python
print(
    pl_pokemon
    .select(['Attack', 'Defense'])
    .corr()
)

# 0.438687 -> medium positive correlation
```

```python
(
    pln.ggplot(data = pl_pokemon)
    + pln.theme_gray()
    + pln.geom_point(mapping = pln.aes(x = 'Attack', y = 'Defense', color = 'Generation'))
    + pln.geom_smooth(mapping = pln.aes(x = 'Attack', y = 'Defense'), method = 'lm')
    + pln.ggtitle("Defense ~ Attack")
    + pln.theme(
        plot_title = pln.element_text(color = "#268026", size = 20, weight = "bold", ha = 0.5),
        axis_title_x = pln.element_text(color = "#8a4412", size = 15, weight = "bold", margin = {'t':15}),
        axis_text_x = pln.element_text(size = 10, weight = "bold"),
        axis_title_y = pln.element_text(color = "#8a4412", size = 15, weight = "bold", margin = {'r':15}),
        axis_text_y = pln.element_text(size = 10, weight = "bold"),
        legend_title = pln.element_text(size = 10, weight = "bold"),
        legend_text = pln.element_text(size = 8, weight = "bold")
    )
)
```

# <span style="color:brown;">Chi2 test for categorical variables</span>

```python
lgnd_gen_table = (
    pl_pokemon
    .group_by(['Legendary', 'Generation'])
    .agg(Count = pl.len())
    .pivot(index = 'Legendary', on = 'Generation', values = 'Count')
)

print(lgnd_gen_table)
```

```python
chi2, p_value, dof, expected_freq = stats.chi2_contingency(lgnd_gen_table)

print(f'Chi2 statistics: {chi2}\n')
print(f'P_value: {p_value}') # 0.0019107013484212193

# There is a relationship between Generation and Legendary
# Meaning, the Generation variable affects the Legendary variable
# (as we can see, pokemons at higher generations have higher chance to be a legendary pokemon)
```

# <span style="color:brown;">Two independent samples test</span>

```python
(
    pl_pokemon
    .filter(c('Legendary') == True)
    .pipe(lambda f: stats.shapiro(f['Attack']))
)
```

```python
(
    pl_pokemon
    .filter(c('Legendary') == False)
    .pipe(lambda f: stats.shapiro(f['Attack']))
)
```

```python
stats.mannwhitneyu(
    x = pl_pokemon.filter(c('Legendary') == True).get_column('Attack'),
    y = pl_pokemon.filter(c('Legendary') == False).get_column('Attack')
)

# pvalue=np.float64(3.4625666210159153e-19)
# There is a significant difference between legendary-true attack and legendary-false attack
```
